# Import data to a list

```python
# import data
from sys import getsizeof
from heapq import *

nums = open("algo1-programming_prob-2sum.txt").read().split()

nums = [int(num) for num in nums]

nums[-10:-1]
```

```python
import sys, threading
sys.setrecursionlimit(800000)
threading.stack_size(67108864)


nums = set(nums)
# t = 48270757984
cnt = 0

def main(s, t):
    global cnt
    
    for t in range(s, t):
        for num in nums:       
            if (t - num) != num and (t-num) in nums:
                cnt += 1
                print(t, cnt)
                break


low = [i for i in range(-1000, 10001, 10)]
thread = threading.Thread(target=main, args=(-10, 10))
thread.start()
```

```python
low = [i for i in range(-10000, 10001, 1000)]
bound = [(low[i], low[i+1]) for i in range(len(low)-1)]
bound[-1] = (bound[-1][0], bound[-1][1] + 1)
threads = []
cnt = 0
for s, t in bound:
    thread = threading.Thread(target=main, args=(s, t))
    thread.start()
    threads.append(thread)
    
[thread.join() for thread in threads]
```
