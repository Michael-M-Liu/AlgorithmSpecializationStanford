Question 
Download the following text file:

kargerMinCut.txt

The file contains the adjacency list representation of a simple undirected graph. There are 200 vertices labeled 1 to 200. The first column in the file represents the vertex label, and the particular row (other entries except the first column) tells all the vertices that the vertex is adjacent to. So for example, the 6th row looks like : "6 155 56 52 120 ......". This just means that the vertex with label 6 is adjacent to (i.e., shares an edge with) the vertices with labels 155,56,52,120,......,etc

Your task is to code up and run the randomized contraction algorithm for the min cut problem and use it on the above graph to compute the min cut. (HINT: Note that you'll have to figure out an implementation of edge contractions. Initially, you might want to do this naively, creating a new graph from the old every time there's an edge contraction. But you should also think about more efficient implementations.) (WARNING: As per the video lectures, please make sure to run the algorithm many times with different random seeds, and remember the smallest cut that you ever find.) Write your numeric answer in the space provided. So e.g., if your answer is 5, just type 5 in the space provided.

```python
import random
from datetime import datetime, timedelta

file = open("kargerMinCut.txt", 'r')
graph = dict()

while True:
    line = file.readline().split()
#     print(line)  
    if line:
        graph[line[0]]=line[1:]
    else:
        break

graph.keys()
```

```python
def random_edge2(graph):
    weights = []
    for key in graph:
        weights.append(len(graph[key]))
#     print(graph.keys(), len(weights))    
    pick_key = random.choices(list(graph.keys()), weights = weights)[0]
    pick_value = random.choice(graph[pick_key])
    return [pick_key, pick_value]

pick = []
for i in range(10):
    pick.append(random_edge2(graph))

print(pick)
```

```python
def contraction(graph):
    time_for_random = time_for_replace = timedelta()
    
    while len(graph) > 2:
        # randomly select one edge from graph to contract
        t0 = datetime.now()
        k, v = random_edge2(graph)
        time_for_random += datetime.now() - t0

        graph[k] = list(filter(lambda x: x != k and x != v, graph[k] + graph[v]))
        
        # replace all the v in the list with k
        t0 = datetime.now()
        for key in graph[v]:
            if key != k:
                graph[key] = [k if e == v else e for e in graph[key]]
                
        time_for_replace += datetime.now() - t0
        
        del graph[v]
        
#     print(time_for_random, time_for_replace)
```

```python
graph_copy = graph.copy()
contraction(graph_copy)
```

```python
min_cut = float('inf')
final_vertex = []
cuts = []
for i in range(200*200*2):
    graph_copy = graph.copy()
#     print(len(graph_copy))
    random.seed(i)
    contraction(graph_copy)
    cut = len(list(graph_copy.values())[0])
    min_cut = min(cut, min_cut)
    final_vertex.append(list(graph_copy.keys()))
    cuts.append(cut)
#     print(cut)
#     print(graph_copy.keys())
#     print(graph_copy.values())
    print(min_cut)
min_cut
```
